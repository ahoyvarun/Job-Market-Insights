# Job Market Insights Project

This project analyzes job listings to extract trends in roles, locations, required skills and salary expectations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import string

df = pd.read_csv("job_listings.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
df['job_title'].value_counts().head(10)
```

#Vizualizing top job titles and locations

```python
job_counts = df['job_title'].value_counts()
print(job_counts.head())

plt.figure(figsize=(10, 5))
sns.barplot(x=job_counts.head(5).values, y=job_counts.head(5).index, palette="Blues_d")
plt.title("Top 5 Job Titles")
plt.xlabel("Number of Listings")
plt.ylabel("Job Title")
plt.tight_layout()
plt.show()


location_counts = df['location'].value_counts()
print(location_counts.head())

plt.figure(figsize=(10, 5))
sns.barplot(x=location_counts.head(5).values, y=location_counts.head(5).index, palette="Greens_d")
plt.title("Top 5 Job Locations")
plt.xlabel("Number of Listings")
plt.ylabel("Location")
plt.tight_layout()
plt.show()
```

#Wordcloud from Descriptions

```python
text = " ".join(df['description']).lower().translate(str.maketrans('', '', string.punctuation))
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Common Keywords in Descriptions")
plt.show()
```

#Skill Frequency Analysis

```python
# List of common technical and analytical skills
skills = [
    "python", "sql", "excel", "machine learning", "deep learning", "tensorflow",
    "power bi", "statistics", "java", "javascript", "react", "hadoop", "spark", "nlp"
]

# Count frequency of each skill in the job descriptions
skill_counts = {}
for skill in skills:
    count = df['description'].str.lower().str.contains(skill).sum()
    skill_counts[skill] = count

# Convert the skill frequency dictionary to a DataFrame
skill_df = pd.DataFrame.from_dict(skill_counts, orient='index', columns=['count'])
skill_df = skill_df.sort_values(by='count', ascending=False)

# Plot the frequencies
plt.figure(figsize=(12, 6))
sns.barplot(x=skill_df['count'], y=skill_df.index, palette="Purples_r")
plt.title("Frequency of Skills in Job Descriptions")
plt.xlabel("Count")
plt.ylabel("Skill")
plt.tight_layout()
plt.show()
```

#Salary Parsing and Analysis

```python
# Remove € symbol and commas, split on "–" or "-"
salary_split = df['salary_estimate'].str.replace('€', '', regex=False)
salary_split = salary_split.str.replace(',', '', regex=False)
salary_split = salary_split.str.split('–|-', expand=True)

# Rename new columns
df['min_salary'] = salary_split[0].astype(float)
df['max_salary'] = salary_split[1].astype(float)

# Calculate average salary
df['avg_salary'] = df[['min_salary', 'max_salary']].mean(axis=1)

# Print a few rows to verify
print(df[['salary_estimate', 'min_salary', 'max_salary', 'avg_salary']].head())

# Group by job title and calculate mean average salary
avg_salary_by_title = df.groupby('job_title')['avg_salary'].mean().sort_values(ascending=False)

# Plot it
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_salary_by_title.values, y=avg_salary_by_title.index, palette="coolwarm")
plt.title("Average Salary by Job Title")
plt.xlabel("Average Salary (€)")
plt.ylabel("Job Title")
plt.tight_layout()
plt.show()
```

### Key Insights:
- Python appeared in 85% of job descriptions
- Data Scientist roles are the most common
- Munich and Berlin are top hiring hubs
